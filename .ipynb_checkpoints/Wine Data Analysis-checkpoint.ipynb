```python
import pandas as pd
import numpy as n
import matplotlib as plot
```

```python
#Load the dataframe
```

```python
filename = "wines/wines.csv"
df = pd.read_csv(filename)
```

```python
#How many wines have been given a rating of 100 points?
```

```python
count = len(df[df['points']==100])
print(count)
```

```python
#What is the name of the most expensive wine?
```

```python
#d1 = df.loc[:,['price','name']]
#df.loc[df['price'] == df['price'].max()]
df.loc[df['price'] == df['price'].max(), 'name'].item()
```

```python
#New column with display of ratings in a scale fromm 0 to 5
#Take existing ratings and convert to 0 to 5 scale
```

```python
new_df = df.copy()
#0-20=1, 21-40=2, 41-60=3, 61-80=4,81-100=5 
new_df.insert(5, "Scale", (new_df['points']/10)/2,True)
```

```python
display(new_df)
```

```python
#Price histogram for wines that cost LESS than $100
```

```python
df_price_less_100 = df.copy()
df_price_less_100 = df.loc[df['price'] <= 100]
count2 = len(df_price_less_100[df_price_less_100['price']<100])
print(count2)
display (df_price_less_100[df_price_less_100['price']<100])
```

```python
df_price_less_100.hist(column='price')
```
